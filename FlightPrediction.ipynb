```python
df['Journey_Day'] = pd.to_datetime(df['Date_of_Journey'],format="%d/%m/%Y").dt.day
```

```python
df.head()
```

```python
df.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
```

```python
df['Dep_Hour'] = pd.to_datetime(df['Dep_Time']).dt.hour
df['Dep_Min'] = pd.to_datetime(df['Dep_Time']).dt.minute
df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Hour'] = pd.to_datetime(df['Arrival_Time']).dt.hour
df['Arrival_Min'] = pd.to_datetime(df['Arrival_Time']).dt.minute
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df.head()
```

```python
duration = list(df['Duration'])

for i in range(len(duration)):
    if len(duration[i].split()) !=2:
        if 'h' in duration[i]:
            duration[i] = duration[i].strip() + " 0m"
        else:
            duration[i] = "0h " + duration[i].strip()

duration_hours = []
duration_minutes = []

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep='h')[0]))
    duration_minutes.append(int(duration[i].split(sep='m')[0].split()[-1]))
```

```python
df['Duration_Hour'] = duration_hours
df['Duration_Min'] = duration_minutes
df.head()
```

```python
df.drop(['Duration'],axis=1,inplace=True)
```

```python
df["Total_Stops"] = np.where(df['Total_Stops'] == 'non-stop','0 stops',df['Total_Stops'])
```

```python
df['Total_Stops'].value_counts()
```

```python
total_stops = list(df['Total_Stops'])
```

```python
stops = []
for i in range(len(total_stops)):
    stops.append(int(total_stops[i].split()[0]))
stops
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
rfr.score(X_test,y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator = rfr, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python

plt.figure(figsize = (8,8))
sns.distplot(y_test-prediction)
plt.show()
```
